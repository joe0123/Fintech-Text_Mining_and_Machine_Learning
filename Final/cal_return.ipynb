```python
import pandas as pd
import csv
import numpy as np
import matplotlib.pyplot as plt
```

```python
buy_data = pd.read_excel('buy_data.xlsx', sheet_name = 'R5-1交易')
#price_data = pd.read_csv('net_value.csv', encoding='utf-8')

with open('net_value3.csv', 'r',encoding="utf-8", errors='ignore') as f:
    price_data = pd.read_csv(f)
```

```python
buy_data.head()
```

```python
price_data.head()
header = list(price_data)[1:]
#print(header)
```

```python
name = buy_data['基金簡稱'][1]
match_data = list(price_data['local'])

for i in range(1, 102):
    #print(match_data[i])
    if name == match_data[i]:
        a=price_data.ix[i]

b = buy_data.loc[1][1:14]

print(a[:4])
a = list(a)[1:]
b = list(b)[1:]

c = pd.DataFrame(
    {'date': header, '野村巴西證券投資信託基金': a, '購買金額':b}
)
```

```python
val = c['購買金額']
cal = []
temp = 0
for i in range(0, 12):
    temp = temp + val[i]
    cal.append(temp)
#print(cal)

c['累積金額'] = cal
```

```python
c.head()
```

```python
price = c['野村巴西證券投資信託基金']
pos = [] #購買部位

for i in range(0, 12):
    temp = val[i] / price[i]
    pos.append(temp)

c['購買部位'] = pos
c.head()
```

```python
cum_pos = [] #累積部位
temp = 0
for i in range(0, 12):
    temp = temp + pos[i]
    cum_pos.append(temp)
    
c['累積部位'] = cum_pos
c.head()
```

```python
account_value = [] #累積帳面價值

for i in range(0, 12):
    temp = cum_pos[i] * price[i]
    account_value.append(temp)

c['累積帳面價值'] = account_value
c.head()
```

```python
act_return = [] #實際購買_報酬
for i in range(0, 12):
    temp = (account_value[i] - cal[i])/cal[i]
    act_return.append(temp)

c['實際購買_報酬'] = act_return
c.head()
```

```python
name = '野村貨幣市場基金'
match_data = list(price_data['local'])

for i in range(1, 102):
    if name == match_data[i]:
        currency = price_data.ix[i]

currency = list(currency)[1:]


c['野村貨幣市場基金'] = currency
c.head()
```

```python
curr_price = c['野村貨幣市場基金']
curr_pos = [] #貨幣型_購買部位

for i in range(0, 12):
    temp = val[i] / curr_price[i]
    curr_pos.append(temp)

c['貨幣型購買部位'] = curr_pos

curr_cum_pos = [] #累積部位
temp = 0
for i in range(0, 12):
    temp = temp + curr_pos[i]
    curr_cum_pos.append(temp)
    
c['貨幣型累積部位'] = curr_cum_pos

curr_account_value = [] #貨幣型累積帳面價值

for i in range(0, 12):
    temp = curr_cum_pos[i] * curr_price[i]
    curr_account_value.append(temp)

c['貨幣型累積帳面價值'] = curr_account_value

curr_return = [] #貨幣購買_報酬
for i in range(0, 12):
    temp = (curr_account_value[i] - cal[i])/cal[i]
    curr_return.append(temp)

c['貨幣型購買_報酬'] = curr_return

c.head()
```
